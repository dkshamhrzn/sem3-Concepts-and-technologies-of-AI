• Solve the following system of Linear equation using Inverse Methods.

2x − 3y + z = −1
x − y + 2z = −3
3x + y − z = 9

{Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}
• Now: solve the above equation using np.linalg.inv function.{Explore more about ”linalg” function
of Numpy}

```python
#1
import numpy as np

# Define matrices A and B
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# 1. Prove A * A^-1 = I
A_inv = np.linalg.inv(A)  # Calculate the inverse of A
identity_matrix = np.dot(A, A_inv)  # Multiply A with its inverse

# Check if the result is approximately an identity matrix
is_identity = np.allclose(identity_matrix, np.eye(A.shape[0]))

# 2. Prove AB != BA
AB = np.dot(A, B)  # Compute A * B
BA = np.dot(B, A)  # Compute B * A
are_not_equal = not np.array_equal(AB, BA)  # Compare the two results

# 3. Prove (AB)^T = B^T * A^T
AB_T = np.transpose(AB)  # Compute the transpose of AB
BT_AT = np.dot(np.transpose(B), np.transpose(A))  # Compute B^T * A^T
is_transpose_equal = np.allclose(AB_T, BT_AT)

# Print results
print("1. A * A^-1 = I:")
print(identity_matrix)
print("Is A * A^-1 approximately the identity matrix?", is_identity)

print("\n2. AB != BA:")
print("AB:\n", AB)
print("BA:\n", BA)
print("Are AB and BA unequal?", are_not_equal)

print("\n3. (AB)^T = B^T * A^T:")
print("(AB)^T:\n", AB_T)
print("B^T * A^T:\n", BT_AT)
print("Is (AB)^T equal to B^T * A^T?", is_transpose_equal)


#2

# Define the coefficient matrix A and the constant vector B
A = np.array([[2, -3, 1],
              [1, -1, 2],
              [3, 1, -1]])
B = np.array([-1, -3, 9])

# Compute the inverse of A
A_inv = np.linalg.inv(A)

# Solve for X using the inverse method
X = np.dot(A_inv, B)

# Verify the solution by calculating A * X
verification = np.dot(A, X)

# Display results
print("Matrix A:\n", A)
print("Vector B:\n", B)
print("\nSolution Vector X (x, y, z):\n", X)
print("\nVerification (A * X):\n", verification)
print("Does A * X approximately equal B?", np.allclose(verification, B))
```

In this exercise, you will compare the performance and implementation of operations using plain Python
lists (arrays) and NumPy arrays. Follow the instructions:
1. Element-wise Addition:
• Using Python Lists, perform element-wise addition of two lists of size 1, 000, 000. Measure
and Print the time taken for this operation.
• Using Numpy Arrays, Repeat the calculation and measure and print the time taken for this

• Using Python Lists, perform element-wise multiplication of two lists of size 1, 000, 000. Mea-

• Using Numpy Arrays, Repeat the calculation and measure and print the time taken for this
operation.
3. Dot Product
• Using Python Lists, compute the dot product of two lists of size 1, 000, 000. Measure and
Print the time taken for this operation.
• Using Numpy Arrays, Repeat the calculation and measure and print the time taken for this
operation.
4. Matrix Multiplication
• Using Python lists, perform matrix multiplication of two matrices of size 1000x1000. Measure
and print the time taken for this operation.

• Using NumPy arrays, perform matrix multiplication of two matrices of size 1000x1000. Mea-
sure and print the time taken for this operation.

```python
import numpy as np
import time

# size of list
list_size = 1_000_000
matrix_size = 1000

# Create lists and arrays
list1 = [i for i in range(list_size)]
list2 = [i for i in range(list_size)]

array1 = np.array(list1)
array2 = np.array(list2)

# 1. Element-wise Addition
# using Python lists
start_time = time.time()
result_list_add = [list1[i] + list2[i] for i in range(list_size)]
list_add_time = time.time() - start_time

# using NumPy arrays
start_time = time.time()
result_array_add = array1 + array2
array_add_time = time.time() - start_time

print(f"Element-wise Addition: Python Lists = {list_add_time:.6f}s, NumPy Arrays = {array_add_time:.6f}s")

# 2. Element-wise Multiplication
# using Python lists
start_time = time.time()
result_list_mul = [list1[i] * list2[i] for i in range(list_size)]
list_mul_time = time.time() - start_time

# using NumPy arrays
start_time = time.time()
result_array_mul = array1 * array2
array_mul_time = time.time() - start_time

print(f"Element-wise Multiplication: Python Lists = {list_mul_time:.6f}s, NumPy Arrays = {array_mul_time:.6f}s")

# 3. Dot Product
# using Python lists
start_time = time.time()
result_list_dot = sum(list1[i] * list2[i] for i in range(list_size))
list_dot_time = time.time() - start_time

# using NumPy arrays
start_time = time.time()
result_array_dot = np.dot(array1, array2)
array_dot_time = time.time() - start_time

print(f"Dot Product: Python Lists = {list_dot_time:.6f}s, NumPy Arrays = {array_dot_time:.6f}s")

# 4. Matrix Multiplication
# Create matrices
matrix1 = [[i for i in range(matrix_size)] for _ in range(matrix_size)]
matrix2 = [[i for i in range(matrix_size)] for _ in range(matrix_size)]

np_matrix1 = np.array(matrix1)
np_matrix2 = np.array(matrix2)

# Python lists
start_time = time.time()
result_matrix = [[sum(a * b for a, b in zip(matrix1_row, matrix2_col)) for matrix2_col in zip(*matrix2)] for matrix1_row in matrix1]
matrix_mul_time = time.time() - start_time

# NumPy arrays
start_time = time.time()
result_np_matrix = np.dot(np_matrix1, np_matrix2)
np_matrix_mul_time = time.time() - start_time

print(f"Matrix Multiplication: Python Lists = {matrix_mul_time:.6f}s, NumPy Arrays = {np_matrix_mul_time:.6f}s")
```
