1. Calculate the total Fare paid across all passengers in the Titanic dataset.

```python
total_fare=df['Fare'].sum()
print("total fare:",total_fare)
```

2. Subset for passengers in first class (Pclass is 1) and calculate their total fare.

```python
first_class_passengers = df[df['Pclass'] == 1]
print(first_class_passengers)
total_fare_first_class = first_class_passengers['Fare'].sum()
print("Total fare for passengers in first class:", total_fare_first_class)
```

3. Do the same for second class (Pclass is 2) and third class (Pclass is 3).

```python
#for second class passenger
second_class_passenger=df[df['Pclass'] == 2]
print("second class passenger subset created")
total_fare_second_class=second_class_passenger['Fare'].sum()
print("total fare for passengers in second class:",total_fare_second_class)
#for third class passenger
third_class_passenger=df[df['Pclass'] == 3]
print("third class passenger subset created")
total_third_class=third_class_passenger['Fare'].sum()
print("total fare for passengers in third class:",total_third_class)
```

4. Combine the fare totals from first, second, and third classes into a list.

```python
total_fare_list=[total_fare_first_class,total_fare_second_class,total_third_class]
print("total fare list:",total_fare_list)
```

5. Divide the totals for each class by the overall total fare to get the proportion of fare revenue by class.

```python
firstclass_proportion=total_fare_first_class/total_fare
secondclass_proportion=total_fare_second_class/total_fare
thirdclass_proportion=total_third_class/total_fare
print("proportion of fare revenue by class:",firstclass_proportion,secondclass_proportion,thirdclass_proportion)
```

What percent of the total number of passengers on the Titanic belonged to each age group
(e.g., child, adult, senior)?

1. Create a new column, age group, that categorizes passengers into "child" (age < 18), "adult" (age
18{64), and "senior" (age 65 and above).

```python
def age_group(age):
    if age < 18:
        return 'child'
    elif age >= 18 and age < 65:
        return 'adult'
    else:
        return 'senior'

df['Age_group'] = df['Age'].apply(age_group)
df[['Name', 'Age', 'Age_group']]
```

2. Calculate the total number of passengers on the Titanic.

```python
total_passengers = len(df)
print("Total number of passengers:", total_passengers)
```

3. Count the number of passengers in each age group.

```python
age_group_counts = df['Age_group'].value_counts()
print(age_group_counts)
```

4. Divide the count of each age group by the total number of passengers to get the proportion of passengers
in each age group.

```python
age_group_proportion = age_group_counts / total_passengers
print(age_group_proportion)
```

5. Display the proportion as a percentage.

```python
age_group_percentage = age_group_proportion * 100
print(age_group_percentage)
```
