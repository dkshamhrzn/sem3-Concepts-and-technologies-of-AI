## Task - 3 - Outlier Detection:

Tasks:
1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
2. Define outliers using the 1.5 × IQR rule.
3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color.
4. Discuss the characteristics of these outliers and their potential impact on regional averages.

```python
#q1,q2
# IQR for both 'Score' and 'GDP per Capita'
def calculate_outliers(df, column):
    """
    parameter:dataset,column
    return:lower and upper bound
    description:function to calculate the IQR for both 'Score' and 'GDP per Capita'
    """
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return lower_bound, upper_bound

#identify outliers based on 'Score' and 'GDP per Capita'
score_lower, score_upper = calculate_outliers(sa_df, 'score')
print("lower bound")
gdp_lower, gdp_upper = calculate_outliers(sa_df, 'Log GDP per capita')


#tag countries as outliers
outliers = sa_df[
    (sa_df['score'] < score_lower) |
    (sa_df['score'] > score_upper) |
    (sa_df['Log GDP per capita'] < gdp_lower) |
    (sa_df['Log GDP per capita'] > gdp_upper)
]
print(outliers)

#scatterplot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=sa_df, x='Log GDP per capita', y='score', color='blue', label='Normal')
#outlier plot
sns.scatterplot(data=outliers, x='Log GDP per capita', y='score', color='red', label='Outliers', s=100)
plt.xlabel('Log GDP per capita')
plt.ylabel('score')
plt.title('Outlier Detection in South Asia: GDP per Capita vs. Score')
plt.legend()
plt.show()


print("Outlier Countries based on Score and GDP per Capita:")
print(outliers[['Country name', 'Log GDP per capita', 'score']])
```

we can find one outlier in the lower part of the graph, through which we can assume that this particular outlier is a lower bound outlier.This specific outlier can possibly decrease the regional mean.
